***Data Cleaning***

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
print("\nUnique Continents:", df['continent'].unique())
```

```python
df['2023 population'] = df['2023 population'].astype(int)
print("\nData Types After Cleaning:")
print(df.dtypes)
```

```python
india_data = df[df['country'] == 'IND'][['country', '2023 population']]
print("\nIndia Data:")
print(india_data)
```

***Data preprocessing***

```python

continent_pop = df.groupby('continent')['2023 population'].sum().reset_index()
print("\nTotal Population by Continent:")
print(continent_pop)
```

***Exploratory data analysis***

```python
print("\nSummary Statistics of Population Counts by Continent:")
print(continent_pop['2023 population'].describe())
```

```python
total_world_pop = continent_pop['2023 population'].sum()
print(f"\nTotal World Population (2023): {total_world_pop:,}")
```

***Visualization***

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(continent_pop['continent'], continent_pop['2023 population'], color=['#FF6384', '#36A2EB', '#FFCE56', '#4BC0C0', '#9966FF', '#FF9F40'])
plt.title("Population Distribution by Continent in 2023")
plt.xlabel("Continent")
plt.ylabel("Population")
plt.xticks(rotation=45)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height/1000000000:.1f}B',
             ha='center', va='bottom')
plt.show()
```
